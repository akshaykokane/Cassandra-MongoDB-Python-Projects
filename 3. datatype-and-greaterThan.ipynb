```python
import pymongo
```

```python
client = pymongo.MongoClient('localhost', 27017)
```

```python
db = client['chapter2']
```

```python
collection = db['income']
```

```python
#read headings from file
with open('income_header.txt') as f_in:
    columns_headings = f_in.readline()
```

```python
columns_headings
```

```python
#convert coloumns heading to list

columns_headings_list = columns_headings.split(',')
```

```python
columns_headings_list
```

```python
#read the data row_wise
row_dict_list = list()
with open('income.txt') as f_in:
    for line in f_in:
        #strip \n and commans
        row_list = line.rstrip('\n').split(',')
        #zip together header for creating key:value pair
        row_dict = dict(zip(columns_headings_list, row_list))
        row_dict_list.append(row_dict)
        collection.insert_one(row_dict)
    
```

```python
collection.count()
```

```python
row_dict_list
```

```python
#age is string and not number in above collection
#so need to convert string to interger type

#delete all the earlier data from collection
collection.delete_many({})
```

```python
with open('income.txt') as f_in:
    for line in f_in:
        #strip \n and commans
        row_list = line.rstrip('\n').split(',')
        
        #zip together header for creating key:value pair
        row_dict = dict(zip(columns_headings_list, row_list))
        
        #before converting age to int, we need to make sure that
        #age contains number only (incase of dirty data)
        #so we need to do error handling
        try:
            #convert age to int
            row_dict['age'] = int(row_dict['age'])
            collection.insert_one(row_dict)
        except:
            pass
```

```python
over_35 = collection.find({'age': {"$gt" : 35}})
```
